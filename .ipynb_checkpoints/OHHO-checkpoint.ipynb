# Load data from db

```python
map_fun = '''function(doc) {
        if (doc.landsize != null)
            emit([doc.price, doc.landsize, doc.address, doc.title, doc.park, doc.bed, doc.bath, doc.longitude, doc.latitude, doc.date, doc.property_type], null);
            }'''
results = db.query(map_fun)
```

```python
df = pd.DataFrame(columns=['price', 'landsize', 'address', 'title', 'park', 'bed', 
                           'bath', 'longitude', 'latitude', 'date', 'property_type'])
```

```python
for i, r in enumerate(results):
    df.loc[i] = r.key
```

# Data clean

```python
df.shape
```

```python
df.info()
```

```python
df.head(2)
```

```python
# trim price column
def trimPrice(x):
    num = x.strip().replace('$','').replace(',','')
    num = num.split()[0]
    try: 
        return int(num)
    except:
        return np.nan

df.price = df.price.map(trimPrice)
df.dropna(inplace=True)
```

```python
# trim landsize
def trimLandsize(x):
    x = x.split('\n')[0].replace('m2', '').split()[-1]
    try:
        return int(x)
    except:
        return np.nan
    
df.landsize = df.landsize.map(trimLandsize)
df.dropna(inplace=True)
```

```python
# trim address
def getSuburb(x):
    return ' '.join(x.lower().split(',')[-1].split()[:-2])

df['suburb'] = df.address.map(getSuburb)
```

```python
def trimAddress(x):
    return x.split(',')[0]
df.address = df.address.map(trimAddress)
```

```python
# trim park, bed, bath
def toInt(x):
    try:
        return int(x)
    except:
        return 0
df.park = df.park.map(toInt)
df.bed = df.bed.map(toInt)
df.bath = df.bath.map(toInt)
```

```python
# get geo
df['geo'] = zip(df.longitude, df.latitude)
df.drop(['longitude', 'latitude'], axis=1, inplace=True)
```

```python
# get sold type
def soldType(x):
    return x.split()[:-3][-1]
df['soldtype'] = df.date.map(soldType)
```

```python
# trim date
def trimDate(x):
    return ' '.join(x.split()[-3:])

df['date'] = df.date.map(trimDate)
df.date = pd.to_datetime(df.date)
```

```python
# trim property_type
def trimPropertyType(x):
    tp = x.split('/')[0].split(',')[0].strip()
    if tp == 'House':
        return 'House'
    elif tp == 'Townhouse':
        return 'Townhouse'
    elif tp == 'Apartment':
        return 'Apartment'
    return np.nan

df.property_type = df.property_type.map(trimPropertyType)
```

```python
# calculate distance to melb_cbd
from math import sin, cos, sqrt, atan2, radians
def calcdist(value):
    a, b = value
    lat1 = radians(-37.810817)
    lon1 = radians(144.963135)
    lat2 = radians(float(b))
    lon2 = radians(float(a))

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    
    R = 6373.0
    distance = R * c

    return distance
df['dist2cbd'] = df.geo.map(calcdist)
df.drop('geo', axis=1, inplace=True)
```

```python
data = df.dropna()
data.shape
```

# Linear model

```python
from sklearn.pipeline import Pipeline
from sklearn.pipeline import make_pipeline

from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet, LinearRegression, RidgeCV, LassoCV, ElasticNetCV
```

## Linear regression

```python
X = data[['landsize', 'park', 'bed', 'bath', 'House', 'Townhouse', 'treaty']]
y = data.price.values
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size=0.5, random_state=49)

ss = StandardScaler()
ss = ss.fit(Xtrain)

Xtrain = ss.transform(Xtrain)
Xtest = ss.transform(Xtest)

lr = LinearRegression()
np.mean(cross_val_score(lr.fit(Xtrain, ytrain), Xtest, ytest, cv=10))
```

## Ridge, Lasso, ElasticNet

```python
ridge_alphas = np.logspace(0, 1, 200)
optimal_ridge = RidgeCV(alphas=ridge_alphas, cv=10)
optimal_ridge.fit(Xtrain, ytrain)

ridge = Ridge(alpha=optimal_ridge.alpha_)
np.mean(cross_val_score(ridge, Xtest, ytest, cv=10))
```

```python
optimal_lasso = LassoCV(n_alphas=500, cv=10, verbose=1)
optimal_lasso.fit(Xtrain, ytrain)

lasso = Lasso(alpha=optimal_lasso.alpha_)
np.mean(cross_val_score(lasso, Xtest, ytest, cv=10))
```

```python
l1_ratios = np.linspace(0.01, 1.0, 25)
optimal_enet = ElasticNetCV(l1_ratio=l1_ratios, n_alphas=100, cv=10,verbose=1)
optimal_enet.fit(Xtrain, ytrain)

enet = ElasticNet(alpha=optimal_enet.alpha_, l1_ratio=optimal_enet.l1_ratio_)
np.mean(cross_val_score(enet, Xtest, ytest, cv=10))
```

## Gradient Boosting Reg

```python
from sklearn.ensemble import GradientBoostingRegressor
clf = GradientBoostingRegressor(n_estimators=500, learning_rate=0.01, min_samples_split = 2,
                                 max_depth=4, random_state=0, loss='ls').fit(Xtrain, ytrain)
np.mean(cross_val_score(clf, Xtest, ytest, cv=10))
```

```python
feature_import = pd.DataFrame([clf.feature_importances_], columns=X.columns).T
feature_import.sort_values(0, ascending=False)
```
