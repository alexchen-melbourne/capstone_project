```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime

%matplotlib inline
pd.set_option('display.max_columns', 200)
```

```python
df = pd.read_excel('data/au.xls', sheetname='Data1')
df.head(2)
```

```python
qu = df.iloc[9:, -9:]
```

```python
au = df.iloc[9:, :9]
```

```python
# Residential Property Price Index of eight capital cities
au.columns = ['Sydney', 'Melbourne', 'Brisbane', 'Adelaide', 'Perth', 'Hobart', 'Darwin', 'Canberra', 'average']
```

```python
def s(x):
    return x.strftime('%Y-%m')
au.index = au.index.map(s)
```

```python
au['index'] = range(au.shape[0])
```

```python
au['time'] = au.index
```

```python
fig = plt.figure(figsize=(20,10))
fig.suptitle('Australia Housing Index', fontsize=20)

NUM_COLORS = 9
cm = plt.get_cmap('Set1')

ax = plt.axes()
plt.xticks(au['index'], au.time)
plt.xticks(rotation=70)
ax.set_color_cycle([cm(1.*i/NUM_COLORS) for i in range(NUM_COLORS)])

# ax.plot(au['index'], au.Sydney)
# ax.plot(au['index'], au.Melbourne)
# ax.plot(au['index'], au.Brisbane)
# ax.plot(au['index'], au.Adelaide)
# ax.plot(au['index'], au.Perth)
# ax.plot(au['index'], au.Hobart)
# ax.plot(au['index'], au.Darwin)
# ax.plot(au['index'], au.Canberra)
# ax.plot(au['index'], au.average)

ax.plot(au['index'], au.Sydney, color='grey')
ax.plot(au['index'], au.Melbourne, color='grey')
ax.plot(au['index'], au.Brisbane, color='grey')
ax.plot(au['index'], au.Adelaide, color='grey')
ax.plot(au['index'], au.Perth, linewidth=5.0, color='orange')
ax.plot(au['index'], au.Hobart, color='grey')
ax.plot(au['index'], au.Darwin, linewidth=5.0, color='brown')
ax.plot(au['index'], au.Canberra, color='grey')
ax.plot(au['index'], au.average, color='grey')

plt.legend(loc = 2, prop={'size': 16})
```
